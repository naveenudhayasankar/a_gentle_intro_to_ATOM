```python
import numpy as np
import pandas as pd 
from sklearn.datasets import fetch_20newsgroups
from atom import ATOMClassifier
```

```python
# Only pick data from the following categories
cats =  ['alt.atheism', 'sci.med', 'comp.windows.x', 'misc.forsale', 'rec.autos']
X,y = fetch_20newsgroups(return_X_y=True, categories=cats)
```

```python
# Reshape input data as ATOM only accepts two dimensional data
X = np.array(X).reshape(-1,1)
X.shape
```

```python
# Define a classifier and test split
atom = ATOMClassifier(X, y, test_size=0.3, random_state=1)
```

```python
atom.dataset
```

```python
# Cleaning the text removes emails, urls, emojis etc. View the deletions using the drops attribute
atom.textclean()
```

```python
atom.drops
```

```python
atom.corpus[0]
```

```python
# Tokenizing the corpus
atom.tokenize()
```

```python
atom.corpus[0]
```

```python
# Visualize the commonly used words using a word cloud
atom.plot_wordcloud()
```

```python
# Normalize and lemmatize so that all words are converted to a uniform standard, also removes stopwords like a, an, the, is which occur frequently and do not provide any useful information
atom.normalize(stopwords='english', lemmatize=True)
```

```python
atom.corpus[0]
```

```python
# N grams are words which give different meaning when occuring together, identify the frequently occuring n grams
atom.plot_ngrams(ngram=2)
```

```python
# Use the tokenizer to join the bigrams together
#atom.tokenize(bigram_freq=200)
atom.bigrams
```

```python
# Vectorizing using TF-IDF
atom.vectorize(strategy='tfidf')
```

```python
atom.shape
```

```python
atom.dataset.dtypes
```

```python
# Fit a random forest classifier model and evaluate using weighted F1 scores
atom.run(models=['RF'], metric='f1_weighted')
```

```python
atom.evaluate()
```

```python
#Plotting the confusion matrix
atom.plot_confusion_matrix()
```

```python
# Visualize which words were involved in the prediction of the first corpus
atom.decision_plot(index=0, target=atom.predict(0), show=10)
```

```python
#Visualize which words were involved in the classifying the first topic (atheism) across all articles
atom.beeswarm_plot(target=0, show=15)
```
